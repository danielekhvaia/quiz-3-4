```python
# csv ფაილის იმპორტი 
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Earthquakes_database.csv") 
df



```

```python
# შლის დუბლირებულ მონაცემებს სტრიქონში
df = df.drop_duplicates()
df
```

```python
# შლის "Depth Error" სვეტს
df = df.drop(columns = "Depth Error")
df
```

```python
# ID სვეტების დაყოფა ახალ სვეტებად

df_split = df['ID'].str.split(expand=True)

# ახალლი სვეტების დამატება მუდმივად, კოდის გამოძახებისას
df = pd.concat([df, df_split], axis=1)
df
```

```python
# ცარიელი მნიშვნელობების დამუშავება fillna-ს და astype-ს გამოყენებით. fillna-ს შეუძლია ცარიელი მნიშვნელობების შევსება ან შეცვლა, ხოლო astype-ს გამოყენებით მონაცემი გადაგვყავს სასურველ ფორმატში.
column = 'Longitude'
df[column] = df[column].fillna(0)
df[column] = df[column].astype(float)
df
```

```python
# ეს ხაზი განსაზღვრავს სვეტის სახელს 'Depth', რათა შემდგომ მარტივად მოვახდინო მასში არსებული ინფორმაციის კატეგორიზაცია
numeric_field_to_cut = 'Depth'

# bin_edges განსაზღვრავს კატეგორიების ინტერვალებს. ამ შემთხვევაში, 50-ზე ნაკლები ან ტოლი მნიშვნელობები არის "Deep", მნიშვნელობები 50-დან 150-მდე არის "Less Deep" და 150-ზე მეტი მნიშვნელობები არის "Very Deep".
# bin_names განსაზღვრავს სახელებს თითოეული კატეგორიისთვის.

bin_edges = [-float('inf'), 50, 150, float('inf')]
bin_names = ['Deep', 'Less Deep', 'Very Deep']


df['Depth_Category'] = pd.cut(df[numeric_field_to_cut], bins=bin_edges, labels=bin_names, right=False)
df

```

```python
multi_index_df = df.set_index(['Type', 'Location Source'])

# შლის დუბლიკატებს ინდექსიდან
multi_index_df = multi_index_df[~multi_index_df.index.duplicated(keep='first')]

# საშუალო სიდიდის გამოთვლა მაგალითისთვის
multi_index_df['Average Magnitude'] = df.groupby(['Type', 'Location Source'])['Magnitude'].mean()

# df -ის დასაბრუნებლად გამოვიყენე unstack() ფუნქცია
unstacked_df = multi_index_df['Average Magnitude'].unstack()
df
```

```python
# სვეტებს შორის ჯვარედინი ტაბულების შექმნა crosstab() - ის გამოყენებით
cross_table = pd.crosstab(df['Type'], df['Location Source'])
df
```

```python
# Bar Chart - მიწისძვრის ტიპების განაწილება
plt.figure(figsize=(10, 6))
df['Type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('მიწისძვრის ტიპების განაწილება')
plt.xlabel('მიწისძვრის ტიპი')
plt.ylabel('სიხშირე')
plt.show()

```

```python
# Scatter Chart - სიღრმისა და მაგნიტუდის კავშირი
plt.figure(figsize=(10, 6))
plt.scatter(df['Depth'], df['Magnitude'], color='red', alpha=0.5)
plt.title('სიღრმისა და მაგნიტუდის კავშირი')
plt.xlabel('სიღრმე')
plt.ylabel('მაგნიტუდი')
plt.grid(True)
plt.show()
```

```python
# Pie Chart - მიწისძვრის მაგნნიტუდის გადანაწილება
plt.figure(figsize=(8, 8))
df['Magnitude'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('მიწისძვრის მაგნნიტუდის გადანაწილება')
plt.show()

```

```python
# Histogram - ჰისტოგრამა.
plt.figure(figsize=(10, 6))
plt.hist(df['Magnitude'], bins=20, color='orange', edgecolor='black')
plt.title('ჰისტოგრამა: მაგნიტუდის განაწილება')
plt.xlabel('Magnitude')
plt.ylabel('რაოდენობა')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
